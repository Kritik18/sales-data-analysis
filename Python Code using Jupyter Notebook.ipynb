## The data at a glance:

```python
#Creating a Dataframe and reading the sales data csv file. 
#Changing the Data type of Date from text to date.

sales = pd.read_csv(r"C:\Users\Kritik\Downloads\sales_data.csv", parse_dates = ["Date"])
sales.head()
```

## Numerical analysis and visualization

We'll analyze the `Unit_Cost` column:

```python
#Mean value of Unit Cost

sales["Unit_Cost"].mean()
```

```python
#Median value of Unit Cost

sales["Unit_Cost"].median()
```

```python
#Created a box plot to find out the outliers in the data

sales["Unit_Cost"].plot(kind='box', vert =False , figsize = (14,6))
```

```python
#Created a density plot to find out where the values are majorly concentrated

sales['Unit_Cost'].plot(kind='density', figsize=(14,6)) # kde
```

```python
#Finding out the mean and median values of the Unit Cost column

ax = sales["Unit_Cost"].plot(kind= "density", figsize = (14,6))
ax.axvline(sales["Unit_Cost"].mean(), color = "red")
ax.axvline(sales["Unit_Cost"].median(), color = "green")
```

```python
ax = sales["Unit_Cost"].plot(kind = "hist", figsize = (14,6))
ax.set_ylabel("Number of Sales")
ax.set_xlabel("Dollars")
```

## Finding the relationship between the columns

```python
numeric_sales = sales.select_dtypes(include = [float,int])
corr = numeric_sales.corr()
corr
```

```python
fig = plt.figure(figsize = (6,6))
plt.matshow(corr, cmap = 'RdBu',fignum = fig.number)
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns)
```

```python
sales['Revenue_per_Age'].plot(kind = 'density')
```

```python
sales['Revenue_per_Age'].plot(kind = 'hist', figsize = (14,6))
```
